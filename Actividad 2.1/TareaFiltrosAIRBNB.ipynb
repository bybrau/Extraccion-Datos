```python
#Lo mismo para cada uno de los filtros...
df = pd.read_csv('listings.csv')

#Convertir la columna 'host_since' a tipo datetime
df['host_since'] = pd.to_datetime(df['host_since'])

#Filtrar los registros para hosts que se unieron después del 2020
df_filtered = df[df['host_since'].dt.year > 2020]

#Mostrar los registros filtrados
print(df_filtered)
```

```python
df_filtered.to_csv('airbnb_filtrado_2020.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_2020.csv'")
```

```python
df = pd.read_csv('listings.csv')

NombresUnicos = df['name'].unique()[:5]
NombresFiltrados = pd.DataFrame()

for name in NombresUnicos:
    NombresFiltrados = pd.concat([NombresFiltrados, df[df['name'] == name].head(1)])

print(NombresFiltrados)

NombresFiltrados.to_csv('airbnb_filtrado_5nombres.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_5nombres.csv'")
```

```python
#Lo mismo
df = pd.read_csv('listings.csv')

#Filtrar los registros para 'host_response_time' igual a 'within an hour' (lo mismo tambien jaja)
RespuestaDias = df[df['host_response_time'] == 'within an hour']

#Mostrar los registros filtrados
print(RespuestaDias)

RespuestaDias.to_csv('airbnb_filtrado_RespuestaDias.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_RespuestaDias.csv'")
```

```python
#Cargar el archivo .CSV 
df = pd.read_csv('listings.csv')

#Filtrar los registros donde 'las filas' es un múltiplo de 200
MultiplosDoscientos = df[df.index % 200 == 0]

#Mostrar los registros filtrados
print(MultiplosDoscientos)

MultiplosDoscientos.to_csv('airbnb_filtrado_MultiplosDoscientos.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_MultiplosDoscientos.csv'")
```

```python
#Cargar el archivo .CSV
df = pd.read_csv('listings.csv')

#Filtrar los registros para 'property_type' igual a 'Entire home' o 'Private room'
TipoPropiedad = df[df['property_type'].isin(['Entire home', 'Private room'])]

#Imprimir
print(TipoPropiedad)

TipoPropiedad.to_csv('airbnb_filtrado_TipoPropiedad.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_TipoPropiedad.csv'")
```

```python
#Cargar
df = pd.read_csv('listings.csv')

#Filtrar mayor o igual a 4
AlMenosCuatro = df[df['beds'] >= 4]

# <Impresión
print(AlMenosCuatro)

AlMenosCuatro.to_csv('airbnb_filtrado_AlMenosCuatro.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_AlMenosCuatro.csv'")
```

```python
df = pd.read_csv('listings.csv')

#Buscar los registros para 'host_is_superhost' igual a 't' o True
SuperHost = df[df['host_is_superhost'].isin(['t', True])]    

#Imprimir 
print(SuperHost)

SuperHost.to_csv('airbnb_filtrado_SuperHost.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_SuperHost.csv'")
```

```python
#Cargar
df = pd.read_csv('listings.csv')

#Filtrar las filas donde el índice es impar
ListaImpares = df[df.index % 2 != 0]

#Mostrar los resul
print(ListaImpares)

ListaImpares.to_csv('airbnb_filtrado_ListaImpares.csv', index=False)
print("Tabla filtrada guardada como 'airbnb_filtrado_ListaImpares.csv'")
```
